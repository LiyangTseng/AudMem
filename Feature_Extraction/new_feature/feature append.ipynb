#  混放zcr&tagging

```python
tagging = pd.read_csv('./PANN_tagging/result/tagging_df/mapping_tag_df.csv')
print(len(tagging))
tagging.head(3)
```

```python
tagging.info()
```

# append zcr & tagging

```python
data['key'] = data.apply(lambda row: str(row.YT_id)+'_'+str(row.augment_type)+'_'+str(row.segment_idx)+'.wav', 
                       axis = 1)
```

```python
data.head(5)
```

```python
df = pd.merge(data, zcr, left_on='key', right_on='youtube_id', how='left')
df = pd.merge(df, tagging, left_on='key', right_on='youtube_id', how='left')
df.head(3)
```

```python
df = df.drop(['key','youtube_id_x', 'youtube_id_y', 'memorability_score_x', 'memorability_score_y'], axis=1)
df.head(3)
```

```python
df['label'] = df.pop('label')
df
```

```python
df.to_csv('./data_newFeature.csv', index=False)
```
